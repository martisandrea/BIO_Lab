Utilizzo della libreria: leggere delle sequenze

```python
from Bio import SeqIO
from Bio.Seq import Seq

records = list(SeqIO.parse("hemoglobin_sequences.fasta", "fasta"))

for record in records:
    print(f"ID: {record.id}")
    print(f"Sequence: {record.seq}")
    print(f"Length: {len(record.seq)}\n")

check = (len(records[0].seq) == len(records[1].seq))
print(f"Do the sequences have the same length? {check}\n")

for record in records:
    print(f"ID: {record.id}. Number of A in the sequence: {record.seq.count('A')}")
```

Trascrizione (DNA -> RNA) e Traduzione (RNA -> Proteina)

```python
dna = list(SeqIO.parse("hemoglobin_sequences.fasta", "fasta"))

rna = [None] * len(dna)
protein = [None] * len(dna)
for i in range(len(dna)):
    # Transcription
    rna[i] = Seq("")
    rna[i].name = dna[i].name
    rna[i].seq = dna[i].seq.transcribe()
    # Translation
    protein[i] = Seq("")
    protein[i].name = dna[i].name
    protein[i].seq = dna[i].seq.translate(to_stop=True)

    print(f"RNA ID {i}: {rna[i].name}")
    print(f"RNA Sequence: {rna[i].seq}")
    print()
    print(f"Protein ID {i}: {protein[i].name}")
    print(f"Protein Sequence: {protein[i].seq}")
    print("\n")

# Back Transcription
bk_dna_test = Seq("")
bk_dna_test.seq = rna[0].seq.back_transcribe()
print(f"Back Translation of protein {dna[0].name}")
print(f"DNA seq: {dna[0].seq}")
print(f"Back-translated DNA seq: {bk_dna_test.seq}")
```

Domande:
* Quanti amminoacidi contiene la proteina?
* Quanti codoni di Stop ci sono (UAA, UAG, UGA)? (posso cercare nel DNA o nell'RNA)


```python
stop_codons = ['UAA', 'UAG', 'UGA']

for i in range(len(dna)):
    # Risposta 1
    print(f"Protein name: {protein[i].name}")
    print(f"Protein Sequence: {protein[i].seq}")
    print(f"Number of amino acids in the sequence: {len(protein[i].seq)}")

    # Risposta 2
    for codon in stop_codons:
        print(f"Total number of stop codon ({codon}) in the RNA sequence: {rna[i].seq.count(codon)}")
        pos = []
        start = 0
        while True:
            j = rna[i].seq.find(codon, start)
            if j == -1:
                break
            else:
                pos.append(j)
                start = j + 1
            if pos == -1:
                print(f"There is no stop codon ({codon})")
        print(f"Positions of stop codon ({codon}): {pos}")

    print("\n")
           
```

```python
aligner = Align.PairwiseAligner()

# Impostazioni dell'aligner
aligner.mode = 'global'
aligner.match_score = 1
aligner.mismatch_score = -2
aligner.open_gap_score = -2
aligner.extend_gap_score = -2

print(f"Algorithm used: {aligner.algorithm}\n")

# get sequence 1 and sequence 2
seq1 = dna[0].seq
name_seq1 = dna[0].name
seq2 = dna[1].seq
name_seq2 = dna[1].name

print(f"Sequences lenghts:")
print(f"{name_seq1}: {len(seq1)}")
print(f"{name_seq2}: {len(seq2)}")

# Compute the alignment score
score = aligner.score(seq1, seq2)
print(f"Alignment score: {score}") 

# Check the first alignment
alignments = aligner.align(seq1, seq2)

# print the alignments
for alignment in alignments:
    # print(f"Alignment score: {alignment.score}")
    print(f"Alignment length: {alignment.shape[1]}\n")
    print(alignment)
    break
```

Substitution Matrices

```python
from Bio.Align import substitution_matrices
aligner = Align.PairwiseAligner()

# Impostazioni dell'aligner
aligner.mode = 'global'
aligner.match_score = 1
aligner.mismatch_score = -2
aligner.open_gap_score = -2
aligner.extend_gap_score = -2
aligner.substitution_matrix = substitution_matrices.load("BLOSUM62")

seq1 = protein[0].seq
name_seq1 = protein[0].name
seq2 = protein[1].seq
name_seq2 = protein[1].name

print(f"Sequences lenghts:")
print(f"{name_seq1}: {len(seq1)}")
print(f"{name_seq2}: {len(seq2)}")

# Compute the alignment score
score = aligner.score(seq1, seq2)
print(f"Best alignment score: {score}\n") 

# Check the first alignment
alignments = aligner.align(seq1, seq2)

# print the alignments
for alignment in alignments:
    print(f"Alignment score: {alignment.score}")
    print(f"Alignment length: {alignment.shape[1]}\n")
    print(alignment)
    break
```
